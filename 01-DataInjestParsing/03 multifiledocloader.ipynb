```python
from langchain.document_loaders import DirectoryLoader, TextLoader
from pathlib import Path

dirPath = Path('./data/documents/')
dir_loader = DirectoryLoader(
    dirPath, 
    glob="**/*.txt",
    loader_cls=TextLoader,
    loader_kwargs={"encoding": "utf8"},
    show_progress=True
    )

dir_docs = dir_loader.load()

for item in dir_docs:
    print(len(item.page_content))  # Print first 100 characters of each document
    print(item.metadata['source'])  # Print the source metadata
    print(item.page_content[:100])  # Print first 100 characters of each document
```
