```python
from langchain.document_loaders import PyMuPDFLoader
import re
from pathlib import Path

class ProcessPDFViaPyMuPDFLoader:

    def __init__(self, pdfPath : Path, pdfName: str):

        self.pdfPath = pdfPath
        self.pdfName = pdfName

    @property
    def pdfPath(self) -> Path:

        return self.__pdfPath
    
    @pdfPath.setter
    def pdfPath(self, pdfPath : Path):
        
        if not pdfPath.exists():
            pdfPath.mkdir(parents=True, exist_ok=True)

        self.__pdfPath = pdfPath

    @property
    def pdfName(self) -> str:

        return self.__pdfName
    

    @pdfName.setter
    def pdfName(self, pdfName: str):
        self.__pdfName = pdfName

    @property
    def pdfLoader(self) -> list:

        pdf_loader = PyMuPDFLoader(self.pdfPath.joinpath(self.pdfName))
        return pdf_loader.load()
    
    def cleanPDFText(self, text: str) -> str:

        
        text = re.sub(r'\s+', ' ', text)

        return text.strip()
    

path = Path("../data/PDFs")
file = "complexpdf.pdf"
mupdfobj = ProcessPDFViaPyMuPDFLoader(path, file)

print(f"Total Pages are:-  {len(mupdfobj.pdfLoader)}")

for item in mupdfobj.pdfLoader:
    print(f"Page Number is: {item.metadata['page'] + 1} ")
    print(f"Total Char in Page: {len(item.page_content)}")
    print(f"top 500 charectors - {mupdfobj.cleanPDFText(item.page_content[:500])}")


```
