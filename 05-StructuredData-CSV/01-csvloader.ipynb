```python
from langchain.document_loaders import CSVLoader
from pathlib import Path
from datetime import datetime
from pytz import timezone
import re


class CSVFileProcess:

    def __init__(self, pathWithFileName : Path):

        self.filePath = pathWithFileName
        self.csvLoader = CSVLoader(
            pathWithFileName,
            csv_args={"delimiter": ",",
                      "quotechar": '"'}
        )

    @property
    def filePath(self) -> Path:
        return self.__filePath
    

    @filePath.setter
    def filePath(self, filePath : Path):

        if str(filePath).strip() == "" or filePath == None:
            raise AttributeError("File Path is Empty")
        elif not filePath.exists():
            filePath.mkdir(parents=True, exist_ok=True)
        else:
            pass

        self.__filePath = filePath

    @property
    def csvFileLoad(self) -> list:
        return self.csvLoader.load()
    

    def cleanText(self, text) -> str:

        text = re.sub(r'\s+',' ', text)
        return text.strip()
    
    def convertCSVtoChunk(self) -> list:

        docs = self.csvFileLoad
        for item in docs:
            print(item)


path = Path("../data/csv")
fileName = "businesscentral_clov.csv"
filePath = path.joinpath(fileName)

obj = CSVFileProcess(filePath)
obj.convertCSVtoChunk()


```
