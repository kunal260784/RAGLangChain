```python
import pandas as pd
from pathlib import Path
from langchain_core.documents import Document 
from pytz import timezone
from datetime import datetime

tz = timezone('America/New_York')

path = Path('../data/csv')

df = []
for item in path.glob('*.csv'):
    df_item = pd.read_csv(item)
    df_item['filename'] = item
    df.append(df_item)

df = pd.concat(df, ignore_index=True)

# print(df)
fileName = ""
fileIndex = 0
docs = []
for i, row in enumerate(df.iterrows()):

    if fileName != row[1]['filename']:
        fileIndex += 1
        fileName = row[1]['filename']

    page = f"""
                sourcelocation : {row[1].sourcelocation}
                pk_uuid : {row[1].pk_uuid}
                number : {row[1].number}
                firstname : {row[1].firstname}
                lastname : {row[1].lastname}
                RoleGroup : {row[1].RoleGroup}
                dept: {row[1].dept}
            """
    
    metadata = {
                "createdBy" : f"{row[1].firstname} {row[1].lastname}",
                "createdDate" : datetime.now(tz=tz).strftime('%Y-%m-%d %H:%M:%S'),
                "Description": row[1].filename,
                "chunkNumber": i+1,
                "fileIndex": fileIndex
    }
    doc = Document(
        page_content=page,
        metadata = metadata
    )   
    docs.append(doc) 


for item in docs:
    print(item.metadata)


```
